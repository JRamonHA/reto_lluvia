```python
import pandas as pd
import glob as glob
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
archivos = glob.glob('../data/*.csv')
archivos = [Path(archivo) for archivo in archivos]

EDO = 'GRO'
f = archivos[0]

def importa_lluvia(f, EDO="GRO"):
    data = pd.read_csv(f, encoding="iso-8859-1")
    data = data[data.EDO==EDO]
    data.columns = ["LON", "LAT", "EDO", "CLAVE", "ESTACION", "LLUVIA"]
    year = f.name[:4]
    mes = f.name[4:6]
    dia = f.name[6:8]
    fecha = pd.to_datetime(f'{year}-{mes}')
    fecha = pd.Timestamp(fecha)
    fecha = [fecha] * len(data)
    data.index = fecha
    data.index
    return data[["EDO", "CLAVE", "ESTACION", "LLUVIA"]]

importa_lluvia(f)
```
